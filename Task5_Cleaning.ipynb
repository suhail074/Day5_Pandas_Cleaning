```python
import pandas as pd
import numpy as np
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0,18,60,100],
    labels=["Child","Adult","Senior"]
)
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

### Final Result

- Missing values handled
- Duplicates removed
- New Age_Group column created
- Cleaned dataset saved successfully
